```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import numpy as np
import os
import tensorflow as tf
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
import pandas_market_calendars as mcal
from datetime import timedelta

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define paths
# test_path = "/kaggle/input/stock-price-prediction-challenge/test"
# indices_path = "/kaggle/input/stock-price-prediction-challenge/train/indices"
test_path = "test"
indices_path = "train/indices"

print("Loading and preparing data...")

# Load test data with enhanced naming
test_dfs = []
for i in range(1, 6):
    df = pd.read_csv(os.path.join(test_path, f"test_{i}.csv"))
    df = df.rename(columns={
        "Returns": f"return_{i}",
        "Close": f"close_{i}",
        "Open": f"open_{i}",
        "High": f"high_{i}",
        "Low": f"low_{i}",
        "Adjusted": f"adjusted_{i}",
        "Volume": f"volume_{i}"
    })
    test_dfs.append(df)

# Merge all stock data
merged_df = test_dfs[0][['Date', f'return_1', f'open_1', f'high_1', f'low_1', f'close_1', f'adjusted_1', f'volume_1']]
for i in range(1, 5):
    df_i = test_dfs[i][['Date', f'return_{i+1}', f'open_{i+1}', f'high_{i+1}', f'low_{i+1}', f'close_{i+1}', f'adjusted_{i+1}', f'volume_{i+1}']]
    merged_df = pd.merge(merged_df, df_i, on='Date', how='outer')

# Add market indices data
indices = {
    "dj": "Dow_Jones.csv",
    "nasdaq": "NASDAQ.csv",
    "SP500": "SP500.csv"
}

for key, filename in indices.items():
    index_df = pd.read_csv(os.path.join(indices_path, filename))
    # Add more columns from indices
    index_df = index_df.rename(columns={
        "Returns": f"returns_{key}",
        "Close": f"close_{key}",
        "Open": f"open_{key}",
        "High": f"high_{key}",
        "Low": f"low_{key}",
        "Adjusted": f"adjusted_{key}",
        "Volume": f"volume_{key}"
    })
    # Merge more informative columns
    merged_df = pd.merge(
        merged_df, 
        index_df[['Date', f'returns_{key}', f'close_{key}', f'open_{key}', f'high_{key}', f'low_{key}']], 
        on='Date', 
        how='left'
    )

# Convert date and sort
merged_df['Date'] = pd.to_datetime(merged_df['Date'])
merged_df = merged_df.sort_values('Date').reset_index(drop=True)

print("Creating advanced technical indicators...")

# Enhanced feature engineering
for i in range(1, 6):
    # Moving averages with different windows
    for window in [5, 10, 20, 50]:
        merged_df[f'ma{window}_{i}'] = merged_df[f'close_{i}'].rolling(window=window).mean()
    
    # Exponential moving averages
    for window in [5, 10, 20]:
        merged_df[f'ema{window}_{i}'] = merged_df[f'close_{i}'].ewm(span=window, adjust=False).mean()
    
    # Price envelopes
    merged_df[f'envelope_upper_{i}'] = merged_df[f'ma10_{i}'] * 1.05
    merged_df[f'envelope_lower_{i}'] = merged_df[f'ma10_{i}'] * 0.95
    
    # Rate of change for different periods
    for period in [5, 10, 20]:
        merged_df[f'roc{period}_{i}'] = merged_df[f'close_{i}'].pct_change(periods=period)
    
    # Volatility indicators
    merged_df[f'volatility_{i}'] = merged_df[f'return_{i}'].rolling(window=20).std()
    
    # Price range indicators
    merged_df[f'high_low_ratio_{i}'] = merged_df[f'high_{i}'] / merged_df[f'low_{i}']
    merged_df[f'close_open_ratio_{i}'] = merged_df[f'close_{i}'] / merged_df[f'open_{i}']
    
    # Volume indicators
    merged_df[f'volume_ma10_{i}'] = merged_df[f'volume_{i}'].rolling(window=10).mean()
    merged_df[f'volume_ratio_{i}'] = merged_df[f'volume_{i}'] / merged_df[f'volume_ma10_{i}']
    
    # Relative Strength Index (RSI)
    delta = merged_df[f'close_{i}'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    merged_df[f'rsi_{i}'] = 100 - (100 / (1 + rs))
    
    # MACD
    merged_df[f'ema12_{i}'] = merged_df[f'close_{i}'].ewm(span=12, adjust=False).mean()
    merged_df[f'ema26_{i}'] = merged_df[f'close_{i}'].ewm(span=26, adjust=False).mean()
    merged_df[f'macd_{i}'] = merged_df[f'ema12_{i}'] - merged_df[f'ema26_{i}']
    merged_df[f'macd_signal_{i}'] = merged_df[f'macd_{i}'].ewm(span=9, adjust=False).mean()
    merged_df[f'macd_hist_{i}'] = merged_df[f'macd_{i}'] - merged_df[f'macd_signal_{i}']
    
    # Beta calculation with different windows
    for window in [10, 20, 30]:
        for index_key in ['dj', 'nasdaq', 'SP500']:
            returns_index_col = f'returns_{index_key}'
            
            # Skip if index data is missing
            if returns_index_col not in merged_df.columns:
                continue
                
            rolling_cov = merged_df[[f'return_{i}', returns_index_col]].rolling(window=window).cov()
            
            cov = rolling_cov.loc[
                rolling_cov.index.get_level_values(1) == returns_index_col,
                f'return_{i}'
            ].reset_index(drop=True)
            
            var = merged_df[returns_index_col].rolling(window=window).var()
            
            # Avoid division by zero
            merged_df[f'beta_{i}_{index_key}_{window}'] = cov / var.replace(0, np.nan)

# Add cross-asset correlations - FIXED VERSION
for i in range(1, 6):
    for j in range(i+1, 6):
        # Calculate rolling correlation
        corr_df = merged_df[[f'return_{i}', f'return_{j}']].rolling(window=20).corr()
        # Extract only the correlation between the two returns (not autocorrelations)
        corr_values = corr_df.loc[corr_df.index.get_level_values(1) == f'return_{j}', f'return_{i}']
        # Reset index and assign to new column
        merged_df[f'corr_{i}_{j}'] = corr_values.reset_index(drop=True)

# Fill NaN values with appropriate methods
merged_df = merged_df.fillna(method='bfill').fillna(method='ffill').fillna(0)

print("Visualizing key features...")

# Visualize key features (keeping your existing visualization code)
fig, axes = plt.subplots(5, 4, figsize=(20, 20), sharex=True)
for i in range(1, 6):
    axes[i-1, 0].plot(merged_df['Date'], merged_df[f'close_{i}'], label='Close', color='blue')
    axes[i-1, 0].set_title(f'Stock {i}  Close')
    axes[i-1, 0].grid()
    
    axes[i-1, 1].plot(merged_df['Date'], merged_df[f'close_{i}'], label='Close', alpha=0.5)
    axes[i-1, 1].plot(merged_df['Date'], merged_df[f'ma10_{i}'], label='MA10', color='orange')
    axes[i-1, 1].plot(merged_df['Date'], merged_df[f'envelope_upper_{i}'], '--', color='green', label='Envelope +5%')
    axes[i-1, 1].plot(merged_df['Date'], merged_df[f'envelope_lower_{i}'], '--', color='red', label='Envelope -5%')
    axes[i-1, 1].set_title(f'Stock {i}  MA10 + Envelopes')
    axes[i-1, 1].legend()
    axes[i-1, 1].grid()
    
    axes[i-1, 2].plot(merged_df['Date'], merged_df[f'roc10_{i}'], label='ROC10', color='purple')
    axes[i-1, 2].set_title(f'Stock {i}  ROC (10 days)')
    axes[i-1, 2].grid()
    
    # Use beta_10 if available, otherwise use another beta column
    beta_col = f'beta_{i}_nasdaq_10'
    if beta_col not in merged_df.columns:
        for alt_index in ['dj', 'SP500']:
            alt_col = f'beta_{i}_{alt_index}_10'
            if alt_col in merged_df.columns:
                beta_col = alt_col
                break
    
    if beta_col in merged_df.columns:
        axes[i-1, 3].plot(merged_df['Date'], merged_df[beta_col], label='Beta', color='black')
        axes[i-1, 3].set_title(f'Stock {i}  Beta to Market')
        axes[i-1, 3].axhline(1, color='gray', linestyle='--', linewidth=0.8)
        axes[i-1, 3].grid()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Correlation analysis
corr_columns = [f'return_{i}' for i in range(1, 6)]
for index in ['dj', 'nasdaq', 'SP500']:
    if f'returns_{index}' in merged_df.columns:
        corr_columns.append(f'returns_{index}')

corr_df = merged_df[corr_columns].dropna()
corr_matrix = corr_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Correlation Matrix of Returns")
plt.show()

# Return time series visualization
plt.figure(figsize=(14, 10))
for i in range(1, 6):
    plt.subplot(5, 1, i)
    plt.plot(merged_df['Date'], merged_df[f'return_{i}'], label=f'Return {i}', color='tab:blue')
    plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
    plt.ylabel('Return')
    plt.title(f'Return {i}')
    plt.grid(True)
    if i == 5:
        plt.xlabel('Date')
plt.tight_layout()
plt.show()

# Return distributions
plt.figure(figsize=(14, 10))
for i in range(1, 6):
    plt.subplot(3, 2, i)
    sns.histplot(merged_df[f'return_{i}'].dropna(), kde=True, bins=50, color='steelblue')
    plt.title(f'Distribution of Return {i}')
    plt.xlabel('Return')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print("Preparing data for LSTM modeling...")

# Enhanced data preparation function
def prepare_data_enhanced(df, return_col, time_step=100, forecast_horizon=10):
    """
    Prepare data for LSTM with enhanced feature selection
    
    Args:
        df: DataFrame containing all features
        return_col: The target return column to predict
        time_step: Lookback window size
        forecast_horizon: Number of days to predict
        
    Returns:
        X: Input sequences with multiple features
        y: Target sequences
    """
    # Select relevant features for this stock
    stock_num = int(return_col.split('_')[1])
    
    # Basic features for this stock
    features = [
        return_col,
        f'close_{stock_num}', 
        f'volume_{stock_num}',
        f'high_low_ratio_{stock_num}',
        f'rsi_{stock_num}',
        f'volatility_{stock_num}',
        f'macd_{stock_num}',
        f'volume_ratio_{stock_num}'
    ]
    
    # Add moving averages
    for window in [5, 10, 20, 50]:
        ma_col = f'ma{window}_{stock_num}'
        if ma_col in df.columns:
            features.append(ma_col)
    
    # Add market index returns
    for index in ['dj', 'nasdaq', 'SP500']:
        if f'returns_{index}' in df.columns:
            features.append(f'returns_{index}')
    
    # Add beta to indices
    for index in ['nasdaq', 'SP500', 'dj']:
        beta_col = f'beta_{stock_num}_{index}_10'
        if beta_col in df.columns:
            features.append(beta_col)
    
    # Create feature dataframe - ensure all columns exist
    features = [f for f in features if f in df.columns]
    feature_df = df[features].copy()
    
    # Scale features - use RobustScaler for better handling of outliers
    scaler_X = RobustScaler()
    feature_values = feature_df.values
    scaled_features = scaler_X.fit_transform(feature_values)
    
    # Scale target separately
    scaler_y = MinMaxScaler(feature_range=(-1, 1))
    target_values = df[return_col].values.reshape(-1, 1)
    scaler_y.fit(target_values)
    
    # Create sequences
    X, y = [], []
    for i in range(time_step, len(scaled_features) - forecast_horizon + 1):
        # For X, use all features
        X.append(scaled_features[i - time_step:i])
        
        # For y, use only the target return column for the next forecast_horizon days
        next_returns = df[return_col].iloc[i:i + forecast_horizon].values
        y.append(next_returns)
    
    return np.array(X), np.array(y), scaler_X, scaler_y

# Enhanced LSTM model function
def create_enhanced_lstm_model(input_shape, forecast_horizon=10):
    """
    Create an enhanced LSTM model with bidirectional layers and batch normalization
    
    Args:
        input_shape: Shape of input data (time_step, n_features)
        forecast_horizon: Number of days to predict
        
    Returns:
        model: Compiled Keras model
    """
    model = Sequential()
    
    # First Bidirectional LSTM layer
    model.add(Bidirectional(LSTM(units=100, return_sequences=True, 
                                 activation='tanh', recurrent_activation='sigmoid'),
                           input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Second Bidirectional LSTM layer
    model.add(Bidirectional(LSTM(units=80, return_sequences=True)))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Third LSTM layer
    model.add(LSTM(units=64, return_sequences=False))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    
    # Dense layers
    model.add(Dense(units=32, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    # Output layer
    model.add(Dense(units=forecast_horizon))
    
    # Compile with Adam optimizer with custom learning rate
    optimizer = Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    return model

print("Training models with time series cross-validation...")

# Define parameters
time_step = 120  # Increased lookback window
forecast_horizon = 10
returns = [f'return_{i}' for i in range(1, 6)]
predictions = {}
validation_mses = {}

# Time series cross-validation
tscv = TimeSeriesSplit(n_splits=5)

# Create callbacks for training
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=0.0001,
    verbose=1
)

# Train models for each return series
for return_col in returns:
    print(f"\nTraining model for {return_col}...")
    
    # Prepare data with enhanced features
    X, y, scaler_X, scaler_y = prepare_data_enhanced(
        merged_df, 
        return_col, 
        time_step=time_step, 
        forecast_horizon=forecast_horizon
    )
    
    # Skip if not enough data
    if len(X) < 10:
        print(f"Not enough data for {return_col}, skipping...")
        continue
    
    # Initialize variables to store cross-validation results
    cv_mses = []
    best_model = None
    best_mse = float('inf')
    
    # Perform time series cross-validation
    for train_idx, val_idx in tscv.split(X):
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        
        # Create and train model
        model = create_enhanced_lstm_model(
            input_shape=(X_train.shape[1], X_train.shape[2]),
            forecast_horizon=forecast_horizon
        )
        
        # Train with callbacks
        history = model.fit(
            X_train, y_train,
            epochs=50,  # Increased epochs with early stopping
            batch_size=32,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )
        
        # Evaluate on validation set
        val_pred = model.predict(X_val)
        val_mse = mean_squared_error(y_val, val_pred)
        cv_mses.append(val_mse)
        
        # Save best model
        if val_mse < best_mse:
            best_mse = val_mse
            best_model = model
        
        # Plot training history
        plt.figure(figsize=(10, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'Training History for {return_col} - Fold MSE: {val_mse:.8f}')
        plt.xlabel('Epoch')
        plt.ylabel('Loss (MSE)')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # Store average validation MSE
    avg_mse = np.mean(cv_mses)
    validation_mses[return_col] = avg_mse
    print(f"Average validation MSE for {return_col}: {avg_mse:.8f}")
    
    # Make predictions for the next 10 days using the best model
    last_sequence = X[-1:] 
    predicted_returns = best_model.predict(last_sequence)[0]
    
    # Store predictions
    predictions[return_col] = predicted_returns

# Print validation MSEs for all stocks
print("\nValidation MSEs for all stocks:")
for return_col, mse in validation_mses.items():
    print(f"{return_col}: {mse:.8f}")

# Visualize predictions
plt.figure(figsize=(14, 10))
for i, return_col in enumerate(returns, start=1):
    if return_col in predictions:
        plt.subplot(3, 2, i)
        plt.plot(range(1, 11), predictions[return_col], label=f'Predicted {return_col}', color='tab:orange')
        plt.title(f'Predicted Next 10 Returns for {return_col}')
        plt.xlabel('Day')
        plt.ylabel('Return')
        plt.grid(True)
plt.tight_layout()
plt.show()

print("Preparing submission file...")

# Get the next 10 trading days after the last date in the data
try:
    # Define specific dates for prediction (March 24, 2025 to April 4, 2025)
    prediction_dates = pd.date_range(start='2025-03-24', end='2025-04-04', freq='B')[:10]
    
    # Create submission dataframe with these specific dates
    submission_df = pd.DataFrame({
        'Date': prediction_dates.strftime('%Y-%m-%d')
    })
except:
    # Fallback: read from sample submission
    print("Using sample submission for dates...")
    sample_submission = pd.read_csv('/kaggle/input/stock-price-prediction-challenge/sample_submission.csv')
    submission_df = pd.DataFrame({'Date': sample_submission['Date']})


# Add predictions to submission dataframe
for i, return_col in enumerate(returns, start=1):
    if return_col in predictions:
        # Ensure column names match the required format
        submission_df[f'Returns_{i}'] = predictions[return_col]
    else:
        # If we don't have predictions for this stock, use zeros
        print(f"Warning: No predictions for {return_col}, using zeros")
        submission_df[f'Returns_{i}'] = [0] * len(submission_df)

# Ensure the submission has the correct format
print("Final submission preview:")
print(submission_df.head())

# Save submission file
submission_df.to_csv('submission.csv', index=False)
print("Submission file saved!")

# Calculate overall expected MSE based on validation results
if validation_mses:
    overall_mse = np.mean(list(validation_mses.values()))
    print(f"\nExpected overall MSE based on validation: {overall_mse:.8f}")

# Additional analysis of predictions
plt.figure(figsize=(12, 8))
for i, return_col in enumerate(returns, start=1):
    if return_col in predictions:
        plt.subplot(5, 1, i)
        plt.plot(range(1, 11), predictions[return_col], 'o-', label=f'Stock {i}')
        plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)
        plt.title(f'Stock {i} Predictions')
        plt.xlabel('Day')
        plt.ylabel('Return')
        plt.legend()
        plt.grid(True)
plt.tight_layout()
plt.show()

# Print prediction statistics
print("\nPrediction Statistics:")
for i, return_col in enumerate(returns, start=1):
    if return_col in predictions:
        pred = predictions[return_col]
        print(f"Stock {i}: Mean={np.mean(pred):.6f}, Std={np.std(pred):.6f}, Min={np.min(pred):.6f}, Max={np.max(pred):.6f}")

# Simpler GRU model for ensemble approach
def create_gru_model(input_shape, forecast_horizon=10):
    model = Sequential()
    model.add(tf.keras.layers.GRU(units=64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(tf.keras.layers.GRU(units=64))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(forecast_horizon))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train ensemble models and average predictions
print("\nTraining ensemble models for improved robustness...")
ensemble_predictions = {}

for return_col in returns:
    if return_col not in predictions:
        continue
        
    print(f"\nTraining ensemble for {return_col}...")
    
    # Get the prepared data
    X, y, scaler_X, scaler_y = prepare_data_enhanced(
        merged_df, 
        return_col, 
        time_step=time_step, 
        forecast_horizon=forecast_horizon
    )
    
    # Skip if not enough data
    if len(X) < 10:
        print(f"Not enough data for {return_col} ensemble, skipping...")
        ensemble_predictions[return_col] = predictions[return_col]
        continue
    
    # Split data
    train_size = int(len(X) * 0.8)
    X_train, X_val = X[:train_size], X[train_size:]
    y_train, y_val = y[:train_size], y[train_size:]
    
    # Train GRU model
    gru_model = create_gru_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        forecast_horizon=forecast_horizon
    )
    
    gru_model.fit(
        X_train, y_train,
        epochs=30,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping],
        verbose=1
    )
    
    # Get predictions from both models
    lstm_pred = predictions[return_col]
    gru_pred = gru_model.predict(X[-1:]).flatten()
    
    # Average the predictions
    ensemble_pred = (lstm_pred + gru_pred) / 2
    ensemble_predictions[return_col] = ensemble_pred

# Create final submission with ensemble predictions
final_submission_df = pd.DataFrame({'Date': prediction_dates.strftime('%Y-%m-%d')})

for i, return_col in enumerate(returns, start=1):
    if return_col in ensemble_predictions:
        final_submission_df[f'Returns_{i}'] = ensemble_predictions[return_col]
    elif return_col in predictions:
        final_submission_df[f'Returns_{i}'] = predictions[return_col]
    else:
        final_submission_df[f'Returns_{i}'] = [0] * len(final_submission_df)

print("\nFinal ensemble submission preview:")
print(final_submission_df.head())

# Save final submission
final_submission_df.to_csv('ensemble_submission.csv', index=False)
print("Ensemble submission file saved!")

print("\nModel training and prediction complete!")

```
